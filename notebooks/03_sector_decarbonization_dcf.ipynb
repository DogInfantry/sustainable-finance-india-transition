```python
import sys; sys.path.insert(0, '..')
import pandas as pd
import matplotlib.pyplot as plt
from src.sector_decarbonization_dcf import SectorDCF, CARBON_PRICE_SCENARIOS
```

## 1. Cement — NPV Sensitivity to Carbon Price

```python
cement = SectorDCF(sector='cement', technology='clinker_substitution')
npv_table = cement.npv_by_carbon_price()
print(npv_table.to_string(index=False))
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
colors = ['#22c55e' if v else '#ef4444' for v in npv_table['viable']]
ax.bar(npv_table['scenario'], npv_table['terminal_npv_inr_cr'], color=colors)
ax.axhline(0, color='black', linewidth=0.8, linestyle='--')
ax.set_title('Cement — Clinker Substitution NPV by Carbon Price Scenario')
ax.set_ylabel('Terminal NPV (INR Crore)')
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

## 2. Steel — Hydrogen DRI vs. Scrap EAF

```python
eaf = SectorDCF(sector='steel', technology='scrap_eaf')
h2_dri = SectorDCF(sector='steel', technology='hydrogen_dri')

eaf_npv = eaf.npv_by_carbon_price()
h2_npv = h2_dri.npv_by_carbon_price()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
for ax, npv_t, title in zip(axes, [eaf_npv, h2_npv], ['Scrap EAF', 'Hydrogen DRI']):
    colors = ['#22c55e' if v else '#ef4444' for v in npv_t['viable']]
    ax.bar(npv_t['scenario'], npv_t['terminal_npv_inr_cr'], color=colors)
    ax.axhline(0, color='black', linewidth=0.8, linestyle='--')
    ax.set_title(f'Steel — {title}')
    ax.set_ylabel('NPV (INR Cr)')
    ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

## Key Observations

- **Cement clinker substitution** is viable at mid carbon prices (~₹3,000/tonne) — low capex, high abatement volume makes this an attractive near-term target.
- **Steel scrap-EAF** is viable across most scenarios given India's growing scrap pool and lower capex vs. hydrogen routes.
- **Hydrogen DRI** requires carbon prices above ₹5,000/tonne to turn NPV-positive — a 2035+ proposition in the base case.
- **Model sensitivity**: Carbon price assumption is the dominant driver. Capex estimates carry ±30% uncertainty in Indian project contexts.
