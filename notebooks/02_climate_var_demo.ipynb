# Portfolio Climate VaR — Demo

Demonstrates the `PortfolioClimateStress` model from `src/portfolio_climate_var.py`.

Runs stress tests across all four climate scenarios and visualises portfolio-level
and sector-level earnings haircuts.

---

```python
import sys; sys.path.insert(0, '..')
import pandas as pd
import matplotlib.pyplot as plt
from src.portfolio_climate_var import PortfolioClimateStress

# Illustrative Nifty 50-like sector weights
holdings = {
    'Financials': 0.28,
    'IT': 0.18,
    'Energy': 0.12,
    'Consumer': 0.11,
    'Materials': 0.10,
    'Healthcare': 0.08,
    'Industrials': 0.07,
    'Utilities': 0.04,
    'Real Estate': 0.01,
    'Communication': 0.01,
}

stress = PortfolioClimateStress(holdings)
print(f'Total weight: {sum(holdings.values()):.2f}')
```

## Single Scenario: Delayed Transition 2030

```python
result = stress.stress_test('transition_2030')
result
```

## All Scenarios — Portfolio Climate VaR Summary

```python
summary = stress.stress_test_all_scenarios()

fig, ax = plt.subplots(figsize=(8, 4))
colors = ['#4ade80' if v > -5 else '#f87171' for v in summary['portfolio_climate_var_pct']]
ax.barh(summary['scenario'], summary['portfolio_climate_var_pct'], color=colors)
ax.axvline(0, color='black', linewidth=0.8)
ax.set_xlabel('Portfolio Climate VaR (%)')
ax.set_title('Portfolio Climate VaR by Scenario')
ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
summary
```
