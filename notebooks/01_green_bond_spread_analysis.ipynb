# Green Bond Spread Analysis — India

This notebook analyses India's green bond issuance history and explores
credit spread dynamics (greenium) vs. duration-matched vanilla bonds.

**Data:** `data/green_bonds/issuance_history_synthetic.csv` (synthetic seed data)  
**For real analysis:** Replace with SEBI registry data or Bloomberg bond export.

---
## Sections
1. Issuance trends by year, sector, issuer type
2. Use-of-proceeds breakdown
3. External review quality distribution
4. Key observations

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

DATA_PATH = Path('../data/green_bonds/issuance_history_synthetic.csv')
df = pd.read_csv(DATA_PATH)
df.head()
```

## 1. Annual Issuance Volume

```python
annual = df.groupby('issuance_year')['amount_inr_cr'].sum().reset_index()
annual.columns = ['Year', 'Total Issuance (INR Cr)']

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(annual['Year'], annual['Total Issuance (INR Cr)'], color='#22c55e', edgecolor='white')
ax.set_title('India Green Bond Issuance — Annual Volume (INR Crore)', fontsize=13)
ax.set_xlabel('Year')
ax.set_ylabel('INR Crore')
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
annual
```

## 2. Issuance by Issuer Type

```python
by_type = df.groupby('issuer_type')['amount_inr_cr'].sum().sort_values(ascending=False)
fig, ax = plt.subplots(figsize=(7, 5))
by_type.plot(kind='bar', ax=ax, color=['#16a34a','#4ade80','#86efac','#bbf7d0'], edgecolor='white')
ax.set_title('Green Bond Issuance by Issuer Type (INR Crore)')
ax.set_xlabel('')
ax.set_ylabel('INR Crore')
ax.grid(axis='y', alpha=0.3)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
by_type.reset_index().rename(columns={'amount_inr_cr': 'Total INR Cr'})
```

## 3. External Review Quality

```python
review = df.groupby('external_review')['amount_inr_cr'].sum()
fig, ax = plt.subplots(figsize=(6, 6))
review.plot(kind='pie', ax=ax, autopct='%1.1f%%', colors=['#16a34a','#4ade80','#d1fae5'])
ax.set_title('Green Bond Issuance by External Review Type')
ax.set_ylabel('')
plt.tight_layout()
plt.show()
```

## 4. Key Observations

- **Sovereign dominates post-2023**: India's Sovereign Green Bond (SGrB) programme launched in FY2023 has rapidly become the largest single issuer by volume.
- **PSU backbone**: NTPC, REC, PFC, and IREDA collectively account for the majority of pre-2023 issuance.
- **Corporate segment is thin**: Private corporate issuance remains limited — regulatory complexity and investor due diligence burden are primary barriers.
- **External review gap**: A meaningful share of issuance (particularly older corporate bonds) has no external review — a structural weakness for market credibility.
