# India Transition Financing Roadmap

This notebook builds an illustrative subsector financing stack for India's clean-energy and transition build-out. Verified narrative claims should always be read with the source ledger in `data/bank_source_ledger.csv`; the annual investment splits are scenario inputs, not market forecasts.

```python
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd

from src.reporting import render_india_transition_roadmap, write_text_report, load_sources
from src.scenarios import (
    build_capital_channel_view,
    build_funding_mix_table,
    build_phase_roadmap,
    add_investment_scenarios,
    load_transition_needs,
)

pd.options.display.max_colwidth = 120
ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
```

```python
transition_needs = load_transition_needs(ROOT / 'data' / 'india_transition_needs.csv')
scenario_view = add_investment_scenarios(transition_needs)
funding_mix = build_funding_mix_table()
capital_channels = build_capital_channel_view(scenario_view)
phase_roadmap = build_phase_roadmap()
sources = load_sources(ROOT / 'data' / 'bank_source_ledger.csv')

scenario_view[['subsector', 'central_case_annual_investment_usd_bn', 'share_of_total_pct', 'roadmap_commentary']]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
scenario_view.plot(
    kind='bar',
    x='subsector',
    y='central_case_annual_investment_usd_bn',
    ax=axes[0],
    legend=False,
    color='#0B6E4F'
)
axes[0].set_title('Illustrative annual investment need by subsector')
axes[0].set_ylabel('USD bn')
axes[0].tick_params(axis='x', rotation=75)

axes[1].pie(
    scenario_view['share_of_total_pct'],
    labels=scenario_view['subsector'],
    autopct='%1.1f%%',
    startangle=120
)
axes[1].set_title('Illustrative share of total annual need')
plt.tight_layout()
```
