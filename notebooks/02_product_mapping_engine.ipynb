# Product Mapping Engine

Rule-based mapper from India transition use case to financing product stack. The notebook keeps use-of-proceeds logic separate from sustainability-linked and transition logic.

```python
from pathlib import Path
import pandas as pd

from src.reporting import load_sources, render_product_mapping_playbook, write_text_report
from src.taxonomy import (
    build_corporate_recommendations,
    build_subsector_recommendations,
    export_product_mapping_table,
    load_corporate_profiles,
    load_product_taxonomy,
    load_transition_needs,
)

pd.options.display.max_colwidth = 120
ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
```

```python
products = load_product_taxonomy(ROOT / 'data' / 'product_taxonomy.csv')
subsectors = load_transition_needs(ROOT / 'data' / 'india_transition_needs.csv')
profiles = load_corporate_profiles(ROOT / 'data' / 'example_corporate_profiles.csv')
sources = load_sources(ROOT / 'data' / 'bank_source_ledger.csv')

mapping_table = export_product_mapping_table(products)
subsector_map = build_subsector_recommendations(subsectors, products, top_n=3)
corporate_map = build_corporate_recommendations(profiles, subsectors, products, top_n=3)
mapping_table
```

```python
playbook = render_product_mapping_playbook(mapping_table, subsector_map, corporate_map, sources)
write_text_report(playbook, ROOT / 'reports' / 'product_mapping_playbook.md')
mapping_table.to_csv(ROOT / 'reports' / 'product_mapping_table.csv', index=False)
print('reports generated')
```
